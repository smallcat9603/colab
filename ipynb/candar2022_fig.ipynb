```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("Book1.csv")
df = df.rename(columns={"d":"Degree"})
df = df.rename(columns={"num_spanning_trees":"# of spanning trees"})
df = df.rename(columns={"num_intersections":"# of intersections"})

print(df)
```

```python
def init_rcParams():

    plt.rcParams['xtick.direction'] = 'in'
    plt.rcParams['ytick.direction'] = 'in'
    plt.rcParams["font.family"] = "Times New Roman"
    plt.rcParams["font.size"] = 20
    plt.rcParams['axes.grid'] = True
    plt.rcParams['figure.figsize'] = [4.0, 4.0]
    plt.rcParams['figure.subplot.top'] = 0.95
    plt.rcParams['figure.subplot.bottom'] = 0.15
    plt.rcParams['figure.subplot.left'] = 0.2
    plt.rcParams['figure.subplot.right'] = 0.95

    # plt.tight_layout()
```

```python
init_rcParams()

df.plot(x="Degree", y="# of spanning trees", marker='o', ylabel="# of spanning trees", legend=False, xticks=range(2,18,2))

plt.ylim([0,250])
plt.savefig("num_span.eps")
plt.savefig("num_span.png")
plt.savefig("num_span.svg")
```

```python
init_rcParams()

df.plot(x="Degree", y="# of intersections", marker='o', ylabel="# of intersections", legend=False, xticks=range(2,18,2))

plt.ylim([0,250])
plt.savefig("num_intersec.eps")
plt.savefig("num_intersec.png")
plt.savefig("num_intersec.svg")
```

```python
df2 = df.copy()
df2 = df2.rename(columns={"hops_max":"max"})
df2 = df2.rename(columns={"hops_avg":"avg"})
df2 = df2.rename(columns={"hops_min":"min"})

print(df2)
```

```python
init_rcParams()

plt.rcParams['figure.subplot.left'] = 0.1
plt.rcParams['figure.figsize'] = [6.0, 4.0]

df2.plot(x="Degree", y=["max", "avg", "min"], kind="bar", ylabel="Average # of hops")
plt.xticks(rotation = 0)
# plt.show()

plt.savefig("ave_hops.eps")
plt.savefig("ave_hops.png")
plt.savefig("ave_hops.svg")
```

```python
df3 = df.copy()
df3 = df3.rename(columns={"reduction_hops_avg":"avg"})
df3 = df3.rename(columns={"reduction_hops_max":"max"})

print(df3)
```

```python
init_rcParams()

plt.rcParams['figure.subplot.left'] = 0.3
plt.rcParams['figure.figsize'] = [4.0, 4.0]

ax = df3.plot.line(x="Degree", y=["max", "avg"], ylabel="Reduction ratio of \n average # of hops", xticks=range(2,18,2), legend=False)

marker = ["s", "o"]
for i, line in enumerate(ax.get_lines()):
    line.set_marker(marker[i])

ax.legend(loc=(0.4,0.3), labelspacing=0.1)
    
plt.xticks(rotation = 0)
# plt.show()

plt.savefig("reduct_hops.eps")
plt.savefig("reduct_hops.png")
plt.savefig("reduct_hops.svg")
```
