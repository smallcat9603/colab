# script to graph data min, max, avg

```python
import matplotlib.pyplot as plt

def boxplot(xlabel=None, ylabel=None, **data):

    plt.boxplot(tuple(data.values()), tick_labels=list(data.keys()))

    if xlabel:
      plt.xlabel(xlabel, fontsize=12)
    if ylabel:
      plt.ylabel(ylabel, fontsize=12)

    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

    ax = plt.gca()
    ax.spines['left'].set_color('black')
    ax.spines['left'].set_linewidth(1)
    ax.spines['right'].set_color('black')
    ax.spines['right'].set_linewidth(1)
    ax.spines['top'].set_color('black')
    ax.spines['top'].set_linewidth(1)
    ax.spines['bottom'].set_color('black')
    ax.spines['bottom'].set_linewidth(1)
    ax.grid(False)

    plt.show()
```

```python
def boxplot_grp(xlabel=None, ylabel=None, legend=[], **data):
    ngroups = len(data)
    nitems = len(legend)
    C = {0: 'C0', 1: 'C1', 2: 'C2', 3: 'C3', 4: 'C4'}
    colors = []

    for n in range(nitems):
        box = plt.boxplot([list(data.values())[group][n] for group in range(ngroups)],
                          positions=[(len(legend)+1)*group+1+n for group in range(ngroups)],
                          patch_artist=True,
                          showmeans=True,
                          medianprops={'color': 'k', 'linewidth': 0},
                          meanprops={'marker':'+',
                                     'markerfacecolor':'k',
                                     'markeredgecolor':'k',
                                     'markersize':5},
                          boxprops={'facecolor': C[n],
                                    'edgecolor': 'grey',
                                    'linewidth': 1})
        colors.append(box['boxes'][0])


    if xlabel:
      plt.xlabel(xlabel, fontsize=12)
    if ylabel:
      plt.ylabel(ylabel, fontsize=12)

    plt.xticks([(nitems+1)/2+(nitems+1)*n for n in range(ngroups)], list(data.keys()), fontsize=12)
    plt.yticks(fontsize=12)

    plt.legend(handles=colors, labels=legend)

    ax = plt.gca()
    ax.spines['left'].set_color('black')
    ax.spines['left'].set_linewidth(1)
    ax.spines['right'].set_color('black')
    ax.spines['right'].set_linewidth(1)
    ax.spines['top'].set_color('black')
    ax.spines['top'].set_linewidth(1)
    ax.spines['bottom'].set_color('black')
    ax.spines['bottom'].set_linewidth(1)
    ax.grid(False)

    plt.tight_layout()
    plt.show()
```

```python
boxplot(x=[2,3,4,5,43,23,1,21,42,54,32],
        y=[2,3,4,5])
```
